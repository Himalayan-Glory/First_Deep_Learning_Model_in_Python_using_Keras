```python
#load dataset
dataset=loadtxt('/home/himalayan/Downloads/indians-diabetes.data.csv',delimiter=',')
```

This is a standard machine learning dataset from the UCI Machine Learning repository.
It describes patient medical record data for Pima Indians and whether they had an onset of diabetes within five years.

```python
#visualizing our dataset
import pandas as pd
df=pd.read_csv('/home/himalayan/Downloads/indians-diabetes.data.csv')
df
```

Input Variables (X):

    1) Number of times pregnant
    2) Plasma glucose concentration a 2 hours in an oral glucose tolerance test
    3) Diastolic blood pressure (mm Hg)
    4) Triceps skin fold thickness (mm)
    5) 2-Hour serum insulin (mu U/ml)
    6) Body mass index (weight in kg/(height in m)^2)
    7) Diabetes pedigree function
    8) Age (years)

Output Variables (y):

    Class variable (0 or 1)


```python
dataset
```

```python
#input 8 columns from 0-7 (i.e.indices sliced by 0:8)
X=dataset[:,0:8]
#output 9th column or 9th variable (i.e. index 8)
Y=dataset[:,8]
```

```python
X
```

```python
Y
```

```python
#defining models
model=Sequential()
    #FIRST HIDDEN LAYER HAS 12 NODES WITH RELU ACTIVATION FXN AND MODEL DATA WITH 8 FEATURES/VARIABLES
model.add(Dense(12,input_dim=8,activation='relu'))
    #SECOND HIDDEN LAYER HAS 8 NODES WITH RELU ACTIVATION FXN
model.add(Dense(8,activation='relu'))
    #OUTPUT LAYER HAS 1 NODES WITH SIGMOID ACTIVATION FXN (since sigmoid function gives binary output, which we are looking for)
model.add(Dense(1,activation='sigmoid'))
```

```python
#compiling the model
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
#fitting the keras model on the dataset
    #Epoch: One pass through all of the rows in the training dataset
    #Batch: One or more samples considered by the model within an epoch before weights are updated
model.fit(X,Y,epochs=100,batch_size=5)
```

```python
#Evaluating the model
accuracy=model.evaluate(X,Y)
```

```python
accuracy_=accuracy*100
print(accuracy_)
```

First values are losses and the second are accuracies

```python
#make prediction
predictions=model.predict(X)
```

Since sigmoid activation has been used here the prediction 
values are obtained between 0 and 1
